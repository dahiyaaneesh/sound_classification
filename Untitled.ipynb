```python
Matrix = []
for fold in range(1,11):
    count_tot     = 0.0
    count_invalid = 0.0
    for label in range(10):
        temp = Meta_data.loc[ (Meta_data["classID"]==label) & (Meta_data["fold"] == fold)][["slice_file_name"]]                     
        for filename in temp.slice_file_name.values:
            
            try:
#                 fs, data = wavfile.read('UrbanSound8K/audio/fold'+str(fold)+"/"+filename)
                data2, samplerate = sf.read('UrbanSound8K/audio/fold'+str(fold)+"/"+filename)
                try:
                    A = np.mean(data2, axis =1)
                except IndexError:
                    A = data2
                A = (A - np.mean(A))/(np.std(A))
                row        = [np.mean(A**i) for i in range(8)]
                row.append(label)
                row.append(fold)
                Matrix.append(row)
                count_tot+=1
            except ValueError:
#                 print("Couldn't read for {} class {} filename {}".format(fold, label, filename))
                count_tot+=1
                count_invalid+=1
    print("Fold {} , unusable :{} , usable :{} , ratio {}".format(fold , count_invalid, count_tot ,count_invalid/count_tot))
        
    
```
